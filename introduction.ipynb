```python
(thresh,binary_img)=cv2.threshold(img_grey,245,255,cv2.THRESH_BINARY)
#(thresh,binary_img)=cv2.threshold(img_grey,128,255,cv2.THRESH_BINARY | cv2.THRESH_OTSU)
#all the values below 128 to 0(black) and 128 to 1(white)
#THresh_otsu- adjust to threshold
#Without thresh_otsu it will only depend on the value given. with thresh_otsu evenif threshold is false it will work
cv2.imshow('Binary image',binary_img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt
plt.imshow(img)
#opencv read image in BGR format matplot display same
#RGB- need to convert to RGB
#opencv has automatic conversion scheme
```

```python
converted_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(converted_img)
```

```python
converted_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB) # if not given triangles are draw again and again
rect=cv2.rectangle(converted_img,pt1=(25,25),pt2=(250,200),color=(0,100,0),thickness=5) 
#if thickness given as negative values will get a filled rectangle
#pt1=top left corner, pt2=botton right corner
plt.imshow(rect)
```

```python
converted_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB) # if given in previous cell triangles are draw again and again
cir=cv2.circle(converted_img,center=(100,100),radius=100,color=(0,255,0),thickness=3)
plt.imshow(cir)
```

```python
converted_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB) # if not given triangles are draw again and again
text=cv2.putText(converted_img,text='Mohanlal',org=(0,150),fontFace=cv2.FONT_HERSHEY_COMPLEX,fontScale=2,color=(255,0,0),thickness=1)
plt.imshow(text)
```
